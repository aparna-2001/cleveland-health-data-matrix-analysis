```python
#Let's do PCA analysis of this data
    #that is principal component analysis


import zipfile
import numpy as np
import pandas as pd

zip_file = "archive.zip"
csv_filename = "heart_cleveland_upload.csv"

with zipfile.ZipFile(zip_file, 'r') as zf:
  with zf.open(csv_filename) as file:
    data_table = pd.read_csv(file)

#let's do PCA
  #step 1 = conversion to matrix
  #step 2 = mean of each column
  #step 3 = subtracting the mean from each column and generating a mean centered matrix
  #step 4 = covariance of the matrix
  #step 5 = Eigen values and Eigen vectors
  #step 6 = ordering the eigen values from largest to smallest
  #step 7 = finding the eigen vector corresponding to the largest Eigen value
  #step 8 = calculating the PCA scores by multiplying mean centered matrix and eigen vectors of the largest eigen value

#step 1 = conversion to matrix;

data_matrix = data_table.select_dtypes(include='number').to_numpy()

#print(data_matrix)

#print(data_table)

#step 2: mean of each column
mean_values = []

for i in range(data_matrix.shape[1]):
  mean_value = np.mean(data_matrix[:,i])
  mean_values.append(mean_value)

mean_values = np.array(mean_values)
#print(mean_values)

#step 3: subtracting the mean from each column

mean_centered_matrix = data_matrix - mean_values
#print(mean_centered_matrix)

#step 4 : finding the covariance of the matrix

covariance_matrix = np.cov(mean_centered_matrix.T)
#print(covariance_matrix)

#step 5 : finding the Eigen values and Eigen vectors

eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
#print(f"Eigen values: {eigen_values}",'\n')
#print(f"Eigen vectors: {eigen_vectors}")


#step 6 : sorting eigen values from largest to smallest

sorted_eigen_indices = np.argsort(eigen_values)[::-1]
sorted_eigen_values = eigen_values[sorted_eigen_indices]
#print(sorted_eigen_values)
#print(sorted_eigen_indices)

#How many largest eigen values are enough?
Total_variance = np.sum(eigen_values) #or sorted_eigen_values. It doesn't matter
variance_ratio = eigen_values/Total_variance
cumulative_variance = np.cumsum(variance_ratio)
#print(cumulative_variance)
#print(variance_ratio)

#here it provides 98% when we reach the third eigen eigen values. so we need to take upto 3 eigen values


#finding the corrsponding eigen vectors of the largest Eigen values, I have taken 3 vectors corresponding to three largest Eigen values

indices_of_large_eigenvalues = sorted_eigen_indices[0:3]

sorted_eigen_vectors = eigen_vectors[:,indices_of_large_eigenvalues]
#print(sorted_eigen_vectors)

#step 7 : calculating the PCA scores by multiplying mean centered matrix and eigen vectors of the largest eigen value

PCA_1_vector = sorted_eigen_vectors[:,0].reshape(-1,1)
PCA_2_vector = sorted_eigen_vectors[:,1].reshape(-1,1)
PCA_3_vector = sorted_eigen_vectors[:,2].reshape(-1,1)

# PCA_1_vector_T = PCA_1_vector.T # This transpose was causing the dimension mismatch
# PCA_2_vector_T = PCA_2_vector.T
# PCA_3_vector_T = PCA_3_vector.T

PCA_1_scores = mean_centered_matrix@PCA_1_vector
PCA_2_scores = mean_centered_matrix@PCA_2_vector
PCA_3_scores = mean_centered_matrix@PCA_3_vector

print(PCA_1_scores)
print(PCA_2_scores)
print(PCA_3_scores)
```
